## **DATA VISUALIZATIONS AND DASHBOARDING**

##### **Objective**

To perform exploratory data analysis on the Superstore Sales dataset and create interactive visualizations and dashboards using Python and Power BI to derive actionable business insights and support data-driven decision making.

### **Importing libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

---

### **Load Superstore dataset**

```python
df = pd.read_csv("../data/cleaned_superstore.csv")
```

---

### **Visualization 1: Sales Trend**

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])

sales_trend = df.groupby(df["Order Date"].dt.month)["Sales"].sum()

plt.figure(figsize=(8,5))
plt.plot(sales_trend,color="red",marker="*")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.savefig("sales_trend.png")
plt.show()
```

---

### **Visualization 2: Sales by Category**

```python
category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(8,5))
category_sales.plot(kind="bar")

plt.title("Sales by Category")

plt.savefig("sales_by_category.png")
plt.show()
```

---

### **Visualization 3: Histogram**

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df["Sales"],
    kde=True
)

plt.title("Sales Distribution")
plt.xlabel("Sales")

plt.show()
```

---

### **Visualization 4: Pair Plot**

```python
sns.pairplot(
    df[["Sales","Profit","Quantity","Discount"]],
    diag_kind="kde"
)

plt.show()
```

---

### **Visualization 5: Profit by Category**

```python
profit_category = df.groupby("Category")["Profit"].sum()

plt.figure(figsize=(8,5))
profit_category.plot(kind="bar")

plt.title("Profit by Category")

plt.savefig("profit_by_category.png")
plt.show()
```

---

### **Visualization 6: Correlation Heatmap**

```python
plt.figure(figsize=(8,5))

sns.heatmap(
    df[["Sales","Profit","Quantity","Discount"]].corr(),
    annot=True
)

plt.title("Correlation Heatmap")

plt.savefig("heatmap.png")
plt.show()
```

---

### **Visualization 7: Sales Distribution**

```python
plt.figure(figsize=(8,5))

plt.hist(df["Sales"], bins=30)

plt.title("Sales Distribution")

plt.savefig("sales_distribution.png")
plt.show()
```

---

### **Visualization 8: Interactive Plotly Chart**

```python
fig = px.bar(
    df,
    x="Category",
    y="Sales",
    color="Category",
    title="Interactive Sales by Category"
)

fig.write_html("interactive_sales.html")

fig.show()
```

---

### **KeyInsights**

1. Sales Performance
Technology category generated the highest sales among all product categories.
Furniture and Office Supplies contributed significantly but had lower sales compared to Technology.
2. Profitability Analysis
Some categories showed high sales but relatively lower profit, indicating the impact of discounts and operational costs.
Technology products produced the highest profit margins.
3. Regional Analysis
Sales were concentrated in a few regions, with certain regions outperforming others in both sales and profit.
Regional performance differences indicate potential opportunities for market expansion.
4. Customer Segment Analysis
The Consumer segment contributed the largest share of overall sales.
Corporate and Home Office segments provided additional revenue streams but with smaller contributions.
5. Product Performance
A small group of products accounted for a significant portion of total sales.
Top-performing products can be prioritized for inventory and marketing strategies.
6. Trend Analysis
Monthly sales trends revealed fluctuations in customer demand throughout the year.
Certain periods experienced higher sales, indicating seasonal purchasing patterns.
