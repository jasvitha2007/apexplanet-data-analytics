### **Step 14: Outlier Detection Using IQR**

##### **Objective**
Detect outliers in the Sales column using the IQR method.

```python
q1=df["Sales"].quantile(0.25)
q3=df["Sales"].quantile(0.75)
IQR=q3-q1
lower=q1 - 1.5 * IQR
upper=q3 + 1.5 * IQR
```

---

### **Step 15: Remove Outliers**

##### **Objective**
Remove extreme sales values from the dataset.

```python
df=df[(df["Sales"]>=lower) & (df["Sales"]<=upper)]
```

---

### **Step 16: Save cleaned Dataset**

##### **Objective**

# **EDA (Exploratory Data Analysis)**

### **Step 17: Sales Distribution Analysis**

##### **Visualization**: **Histogram**

##### **Objective**
Understand the distribution of sales values.

```python
plt.figure(figsize=(9,5))
plt.hist(df["Sales"],bins=25,color="blue")
plt.xlabel="Sales"
plt.ylabel="Frequency"
plt.title("Sales Data Distribution")
plt.show()
```

---

#### **Step 18: Boxplot Analysis using Matplotlib Library**

##### **Visualization: Boxplot**

##### **Objective**
Visualize the spread of sales values in different cities and identify outliers.

```python
c=df[df["City"]=="Los Angeles"]["Sales"]
c1=df[df["City"]=="Costa Mesa"]["Sales"]
c2=df[df["City"]=="Henderson"]["Sales"]
plt.figure(figsize=(8,5))
plt.boxplot([c,c1,c2],label=["Los Angeles","Costa Mesa","Henderson"])
plt.title("Box Plot between sales and city")
plt.show()
```

---

### **Step 19: Boxplot Analysis**

### **Step 20: Category-wise Sales Analysis**

##### **Visualization: Bar Chart**

##### **Objective**
Compare total sales across product categories.

### **Step 23: Monthly Sales Trend**

##### **Visualization: Line Chart**

##### **Objective**
Analyse monthly sales trends

```python
monthly_sales=df.groupby(df["Order Date"].dt.month)["Sales"].sum()
monthly_sales.plot(marker="o",linewidth=2,linestyle=":",color="blue")
plt.title("Monthly Sales Trend")
plt.grid()
plt.show()
```

```python
monthly_sales=df.groupby(df["Order Date"].dt.month)["Sales"].sum()
monthly_sales.plot(marker="*",linewidth=2,linestyle="--",color="red")
plt.title("Monthly Sales Trend")
plt.show()
```

##### **Observation**

Sales fluctuate across different months indicating seasonal trends
