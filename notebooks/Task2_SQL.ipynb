```python
import pandas as pd 
from sqlalchemy import create_engine
```

---

##### **Loading the Cleaned Dataset**
The cleaned dataset from Task 1 is loaded into Pandas.

```python
df=pd.read_csv("../data/cleaned_superstore.csv")
df.head()
```

---

##### **Creating SQLite Database**

The cleaned dataset is stored inside a SQLite database table.

```python
engine=create_engine("sqlite:///sales.db")
df.to_sql(
    "sales",
    engine,
    if_exists="replace",
    index=False
)
```

---

## **SQL Fundamentals**

Basic SQL operations are performed to understand data retrieval and analysis

#### **Practice SELECT**

##### **SELECT Statement**
Retrieve records from the sales table

```python
pd.read_sql("""
            SELECT *
            FROM sales
            LIMIT 5
            """,engine)
```

##### **Observation**

The query successfully retrieves the first five records from the database table.

---

#### **Practice WHERE**

##### **WHERE Clause**
Filter records based on conditions.

```python
pd.read_sql("""
            SELECT [Customer Name]
            FROM sales
            WHERE Sales>50
            """,engine)
```

##### **Observation**

The query retrieves the Customer Name whose sales are greater than 50

---

#### **Practice ORDER BY**

##### **ORDER BY Clause**
Sort records in descending order of sales.

```python
pd.read_sql(""" 
            SELECT *
            FROM sales
            ORDER BY Sales DESC
            LIMIT 10
            """,engine)
```

##### **Observation**

Retrieves the records which are sorted in descending order.

---

#### **Practice LIMIT**

```python
pd.read_sql("""
            SELECT *
            FROM sales
            LIMIT 10
            """,engine)
```

##### **Observation**

Retrieves 10 records from the database

---

#### **Practice GROUP BY**

```python
pd.read_sql(""" 
            SELECT Category,
            SUM(Sales) AS TotalSales
            FROM sales
            GROUP BY Category
            """,engine)
```

##### **Observation**

The GROUP BY clause groups records based on category and calculates total sales for each category. This helps compare category-wise performance.

---

#### **Practice HAVING**

```python
pd.read_sql(""" 
            SELECT Category,
            SUM(Sales) AS TotalSales
            FROM sales
            GROUP BY Category
            HAVING TotalSales > 300000
            """,engine)
```

### Observation

The HAVING clause filters grouped data after aggregation. It displays only categories that satisfy the specified sales threshold.

---

## **Advanced SQL Section**

### **Advanced SQL Concepts**

#### **Practice SubQuery** 

```python
pd.read_sql(""" 
            SELECT *
            FROM sales
            WHERE Sales>
            (
                SELECT AVG(Sales)
                FROM sales
            )
            """,engine)
```

##### **Observation**

The subquery calculates the average sales value, and the outer query displays records with sales above the average. This helps identify high-performing transactions.

---

#### **Practice CTE**

```python
pd.read_sql("""
            WITH Category_sales AS
            (
                SELECT Category,
                SUM(Sales) AS TotalSales
                FROM sales
                GROUP BY Category
            )
            SELECT *
            FROM Category_sales
            """,engine)
```

##### **Observation**

The Common Table Expression (CTE) improves query readability by creating a temporary result set that can be referenced later.

---

#### **Practice ROW_NUMBER**

```python
pd.read_sql(""" 
            SELECT [Product Name],
            Sales,
            ROW_NUMBER() OVER
            (ORDER BY Sales DESC) AS Rowno
            FROM sales
            LIMIT 20
            """,engine)
```

##### **Observation**

ROW_NUMBER assigns a unique sequential number to each row based on sales ranking. It is useful for ranking and reporting purposes.


---

#### **Practice RANK**

```python
pd.read_sql(""" 
            SELECT [Product Name],
            Sales,
            RANK() OVER 
            (ORDER BY Sales DESC) AS Rankno
            FROM sales
            LIMIT 20
            """,engine)
```

##### **Observation**

RANK assigns positions based on sales values. Products with equal sales receive the same rank.

---

#### **Practice LAG**

```python
pd.read_sql(""" 
            SELECT [Order Date],
            Sales,
            LAG(Sales,1) OVER
            (ORDER BY [Order Date]) AS PreviousSales
            FROM sales
            LIMIT 20
            """,engine)
```

##### **Observation**

LAG retrieves the previous row's sales value, enabling comparison between consecutive transactions.

---

#### **Practice LEAD**

```python
pd.read_sql(""" 
            SELECT [Order Date],
            Sales,
            LEAD(Sales,1) OVER
            (ORDER BY [Order Date]) AS NextSales
            FROM sales
            LIMIT 20
            """,engine)
```

##### **Observation**

LEAD retrieves the next row's sales value, making it easier to compare current and future records.

---

#### **Create View**

```python
with engine.connect() as conn:
    conn.exec_driver_sql("DROP VIEW IF EXISTS TopSales")
```

```python
with engine.connect() as conn:
    conn.exec_driver_sql(""" 
    CREATE VIEW TopSales AS
    SELECT *
    FROM sales
    WHERE Sales > 10
""")
```

```python
pd.read_sql(""" 
            SELECT *
            FROM TopSales
            LIMIT 5
            """,engine)
```

##### **Observation**

The view was created successfully and stores frequently used query results. It simplifies access to high-value sales records without repeatedly writing the same SQL query.

# **BUSINESS QUESTIONS SECTION**

#### **1) Rank Products Based on Sales Performance**

```python
pd.read_sql(""" 
            SELECT [Product Name],
            Sales,
            RANK() OVER (ORDER BY Sales DESC) AS
            ProductRank
            FROM sales
            LIMIT 20
            """,engine)
```

##### **Insight**

Ranking products helps identify best-performing items and supports business decision-making

---

#### **2) What are the Top 5 Products by Sales?**

```python
pd.read_sql(""" 
            SELECT [Product Name],
            SUM(Sales) AS TotalSales
            FROM sales
            GROUP BY [Product Name]
            ORDER BY TotalSales DESC
            LIMIT 5
            """,engine)
```

##### **Insight**

The top 5 products contribute a significant portion of total revenue. These products should be prioritized for inventory management and promotional campaigns.

---

#### **3) Which Product Category Generates the Highest Revenue?**

```python
pd.read_sql(""" 
            SELECT Category,
            SUM(Sales) AS TotalSales
            FROM sales
            GROUP BY Category
            ORDER BY TotalSales DESC
            """,engine)
```

##### **Insight**

The category with the highest revenue contributes most to the company's sales performance and should receive continued investment.

---

#### **4) Which Region has the Highest Salary?**

```python
pd.read_sql(""" 
            SELECT Region,
            SUM(Sales) AS TotalSales
            FROM sales
            GROUP BY Region
            ORDER BY TotalSales DESC
            """,engine)
```

##### **Insight**

The highest-performing region can serve as a benchmark for improving sales strategies in other regions.

---

#### **5) Who are the top 10 Customers by Sales?**

```python
pd.read_sql(""" 
            SELECT [Customer Name],
            SUM(Sales) AS TotalSales
            FROM sales
            GROUP BY [Customer Name]
            ORDER BY TotalSales DESC
            LIMIT 10
            """,engine)
```

##### **Insight**

A small number of customers contribute heavily to revenue, highlighting the importance of customers retention.

---

#### **6) What is the monthly sales trend?**

```python
pd.read_sql(""" 
            SELECT strftime('%Y-%m',[Order Date]) AS
            Month,
            SUM(Sales) AS TotalSales
            FROM sales
            GROUP BY Month
            ORDER BY Month
            """,engine)
```

##### **Insight**

Monthly trends help identify seasonal patterns and support future sales forecasting.

---

#### **7) Which Products Generate the Highest Profit?**

```python
pd.read_sql(""" 
            SELECT [Product Name],
            SUM(Profit) AS TotalProfit
            FROM sales
            GROUP BY [Product Name]
            ORDER BY TotalProfit DESC
            LIMIT 10
            """,engine)
```

##### **Insight**

High-Profit products should receive additional marketing attention because they contribute significantly to business growth.

---

#### **8) Which Products generate the lowest Profit?**

```python
pd.read_sql(""" 
            SELECT [Product Name],
            SUM(Profit) AS TotalProfit
            FROM sales
            GROUP BY [Product Name]
            ORDER BY TotalProfit ASC
            LIMIT 10
            """,engine)
```

##### **Insight**

Products with low or negative profits should be reviewed for pricing, discounts, or operational inefficiencies.

---

#### **9) Which Customer Segment Contributes the most sales?**

```python
pd.read_sql(""" 
            SELECT Segment,
            SUM(Sales) AS TotalSales
            FROM sales
            GROUP BY Segment
            ORDER BY TotalSales DESC
            """,engine)
```

##### **Insight**

The most valuable customer segment should be the primary focus for future marketing strategies.

---

#### **10) What is the Average sales value by Category?**

```python
pd.read_sql(""" 
            SELECT Category,
            AVG(Sales) AS AverageSales
            FROM sales
            GROUP BY Category
            """,engine)
```

##### **Insight**

Average sales values help understand customer spending behavior across different product categories.
